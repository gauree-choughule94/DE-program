```python
import numpy as np
import pandas as pd

date1 = pd.date_range(start="2022-01-01", periods=10, freq="D")

data = {
    "Timestamp": date1,
    "value": np.random.randint(10, 100, size=10),
    "score": np.random.rand(10)
}

df = pd.DataFrame(data)
df
```

```python
df['Timestamp'] = df['Timestamp'].dt.tz_localize('UTC')

df['Timestamp']
```

```python
df
```

```python
df['Timestamp'] = df['Timestamp'].dt.tz_convert('Asia/Kolkata')
df
```

```python
df['Timestamp'] = df['Timestamp'].dt.tz_convert('US/Eastern')
df
```

```python
df['Timestamp'] = df['Timestamp'].dt.tz_convert('Canada/Central')
df
```

```python
df['Timestamp'].dt.tz
```

```python
df['utc_offset'] = df['Timestamp'].apply(lambda x: x.utcoffset())
df
```

```python
df['tzname'] = df['Timestamp'].apply(lambda x: x.tzname())
df
```
